```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn import metrics
from sklearn.preprocessing import StandardScaler
import joblib
```

```python
car_df = pd.read_csv('/content/car data.csv')
car_df.head()
```

```python
car_df.shape
```

```python
car_df.info()
```

```python
print(car_df.Fuel_Type.value_counts())
```

```python
replacement_dict = {
    'Fuel_Type': {'Petrol': 0, 'Diesel': 1, 'CNG': 2},
    'Seller_Type': {'Dealer': 0, 'Individual': 1},
    'Transmission': {'Manual': 0, 'Automatic': 1}
}
car_df.replace(replacement_dict, inplace=True)
```

```python
car_df.head()
```

```python
vectorizer = TfidfVectorizer()
vectorizer.fit(car_df['Car_Name'])
Car_Name_vector = vectorizer.transform(car_df['Car_Name'])
joblib.dump(vectorizer, 'vectorizer.pkl')
```

```python
print(Car_Name_vector)
```

```python
X = car_df.drop(['Car_Name','Selling_Price'],axis=1)
Y = car_df['Selling_Price']
```

```python

Car_Name_vector = Car_Name_vector.toarray()
```

```python
type(Car_Name_vector)
```

```python
Car_Name_vector = pd.DataFrame(Car_Name_vector)
```

```python
Car_Name_vector.head()
```

```python
#Car_Name_df = Car_Name_vector.sum(axis=1)
```

```python
#Car_Name_df.head()
```

```python
X = pd.concat([Car_Name_vector,X], axis=1)
```

```python
X.head(10)
```

```python
Y.head()
```

```python
scaler = StandardScaler()
X.columns = X.columns.astype(str)
X_scaled = scaler.fit_transform(X)
joblib.dump(scaler, 'scaler.pkl')
X_scaled = pd.DataFrame(X_scaled, columns=X.columns)
X_scaled.head()
```

```python
Y_scaler = StandardScaler()
Y_scaled = Y_scaler.fit_transform(Y.values.reshape(-1,1))
joblib.dump(Y_scaler, 'Y_scaler.pkl')
Y_scaled = pd.DataFrame(Y_scaled, columns=['Selling_Price'])
Y_scaled.head()
```

```python
X_scaled_train,X_scaled_test,Y_scaled_train,Y_scaled_test = train_test_split(X_scaled,Y_scaled,test_size=0.2,random_state=2)
```

```python
X_scaled_train.shape, X_scaled_test.shape
```

```python
Y_scaled_train.shape, Y_scaled_test.shape
```

```python
linear_model = LinearRegression()
```

```python
linear_model.fit(X_scaled_train,Y_scaled_train)
```

```python
training_prediction = linear_model.predict(X_scaled_train)
```

```python
training_prediction[:20]
```

```python
error_score = metrics.r2_score(Y_scaled_train,training_prediction)
print('Training_Accuracy : ',error_score)
```

```python
plt.figure(figsize=(6,6))
plt.scatter(x=Y_scaled_train,y=training_prediction)
plt.xlabel('Actual Price')
plt.ylabel('Predicted Price')
plt.plot([Y_scaled_train.min(),Y_scaled_train.max()],[Y_scaled_train.min(),Y_scaled_train.max()],'red')
plt.show()
```

```python
test_prediction = linear_model.predict(X_scaled_test)
```

```python
test_prediction[:10]
```

```python
error_score = metrics.r2_score(Y_scaled_test, test_prediction)
print('Test_Accuracy : ',error_score)
```

```python
plt.figure(figsize=(6,6))
plt.scatter(x=Y_scaled_test,y=test_prediction)
plt.xlabel('Actual Price')
plt.ylabel('Predicted Price')
plt.plot([Y_scaled_test.min(),Y_scaled_test.max()],[Y_scaled_test.min(),Y_scaled_test.max()],'red')
plt.show()
```
